# Classical ML baseline research

MFCC/statistical features with logistic regression, SVM, and random forest. Run this notebook after `data/processed/index.csv` exists.

```python
from src.data.annotations import CLASS_NAMES
from src.features.mfcc import build_feature_table
from src.models.ml import build_estimator
from src.utils.metrics import compute_metrics

features, labels, frame = build_feature_table('data/processed/index.csv', 16000, 8.0, 40)
train = frame['split'].eq('train').to_numpy()
test = frame['split'].eq('test').to_numpy()
for name in ['logistic_regression', 'svm', 'random_forest']:
    model = build_estimator(name, 'balanced', 42)
    model.fit(features[train], labels[train])
    pred = model.predict(features[test])
    print(name, compute_metrics(labels[test].tolist(), pred.tolist(), CLASS_NAMES))
```
